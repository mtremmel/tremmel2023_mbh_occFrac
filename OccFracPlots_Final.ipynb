```python
def binomial_to_plot_errors(y_center, bin_err):
    err_lo = bin_err[:,0]
    err_hi = bin_err[:,1]
    bad_lo = np.where(np.isnan(err_lo))
    bad_hi = np.where(np.isnan(err_hi))
    err_lo[bad_lo] = y_center[bad_lo]
    err_hi[bad_hi] = y_center[bad_hi]
    dy_min = y_center - err_lo
    dy_max = err_hi-y_center
    return dy_min,dy_max
```
